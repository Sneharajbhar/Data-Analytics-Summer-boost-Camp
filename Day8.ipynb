```python
df['Grade']=['A','B','C']
df['Marks']=df['Marks']+5
print(df)
```

```python
# Columnwise Drop
df.drop('Grade', axis=1, inplace=True,errors='ignore') # inplace=true means original dataframe changed, if not there then everywhere same named column deleted
# Rowwise drop
df.drop(1,axis=0,inplace=True)
print(df.columns)
```

```python
# sorting
print(df.sort_values(by="Marks",ascending=False))
```

```python
# matploting
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,2,9,16]
plt.plot(x,y,color = 'Blue',linestyle='--',marker='o')
plt.title("line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Bar Chart
x = [1,2,3,4]
y = [1,2,9,16]
plt.bar(x,y,color="Violet")
plt.title("Bar Chart")
plt.show()
```

```python
# horizontal Bar graph 
x = [1,2,3,4]
y = [1,2,9,16]
plt.barh(x,y,color="Violet")
plt.title("Bar Chart horizontal")
plt.show()
```

```python
# Scatter chart
x = [1,2,3,4]
y = [1,2,9,16]
plt.scatter(x,y,color="Violet")
plt.title("Sactter plot Chart")
plt.show()
```

```python
# histograpy
import numpy as np
data = np.random.randn(1000)
plt.hist(data,bins=30,color='Green')
plt.show()
```

```python
# Pie chart
labels = ["Python","Java","C++"]
size = [60,70,20]
plt.pie(size,labels=labels,autopct="%1.1f%%",startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# bar grapy using csv file
df = pd.read_csv('sample.csv')
plt.bar(df['battery_capacity_kWh'], df['top_speed_kmh'], color='green')
plt.xlabel('Model')
plt.ylabel('Top Speed (km/h)')
plt.title("Example: Bar Graph")
plt.show()
```

```python
# matploting
# 
import matplotlib.pyplot as plt
plt.plot(df['number_of_cells'],df['top_speed_kmh'],color = 'Blue',linestyle='--',marker='o')
plt.title("line plot example")
plt.xlabel("Number_of_cells")
plt.ylabel("top_speed_kmh")
plt.grid(True)
plt.show()
```
